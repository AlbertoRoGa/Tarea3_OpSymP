1. Obtener máximos y mínimos absolutos, usando `sympy`, de las funciones dadas en los intervalos dados, y graficar la función en dicho intervalo señalando los puntos máximo y mínimo.

    + $f(x)=2x^3+4x^2+1$ en $[-1.5,\,0.5]$
    + $g(x) = 5x^3-3x+10$ en el intervalo $[-1, \,0]$

```python
import numpy as np
import sympy as sym
sym. init_printing(use_latex='mathjax')
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*x**3 + 4*x**2 + 1
```

```python
sym.var("x")
df = sym.diff(f(x),x)
df
```

```python
xc = sym.solve(df,x)
xc
```

```python
xnum = np.linspace(-1.5,0.5,300)
ynum = f(xnum)
```

```python
plt.figure(figsize=(10,8))
plt.plot(xnum,ynum,"k",lw=2,label="$y = 2x^3 + 4x^2 + 1$")
plt.plot([-4/3],[3.37],"pb",ms=15,label="$(-4/3,3.37)$ Máximo")
plt.plot([0],[1],"pr",ms=15,label="$(0,1)$ Mínimo")
plt.xlabel("$x$",fontsize=15)
plt.ylabel("$y$",fontsize=15)
plt.grid()
plt.legend(loc="best")
plt.show()
```

```python
def g(x):
    return 5*x**3 - 3*x + 10
```

```python
sym.var("x")
dg = sym.diff(g(x),x)
dg
```

```python
xc = sym.solve(dg,x)
xc
```

```python
xnum = np.linspace(-1,0,300)
ynum = g(xnum)
```

```python
import math
plt.figure(figsize=(10,8))
plt.plot(xnum,ynum,"--k",lw=3,label="$y = 5x^3-3x + 10$")
plt.plot([-math.sqrt(5)/5],[10.89],"ob",ms=15,label="$(-\sqrt{5}/5,9.11)$ Máximo")
plt.plot([-1],[8],"or",ms=15,label="$(-1,8)$ Mínimo")
plt.xlabel("$x$",fontsize=15)
plt.ylabel("$y$",fontsize=15)
plt.grid()
plt.legend(loc="best")
plt.show()
```
